```python
# Your code here
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=3,penalty='l1')
model.fit(X_train, y_train)
print(model)
# make predictions
expected = y_test
predicted = model.predict(X_test)
# summarize the fit of the model

from sklearn.metrics import accuracy_score
accuracy_score(expected, predicted)

#стоит l1 регуляризация
```

```python
#l1

from sklearn.linear_model import Lasso
lasso = Lasso().fit(X_train, y_train)

print("Правильность на контрольном наборе: {:.2f}".format(lasso.score(X_test, y_test)))
#почему?
```

```python

#l2
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(X_train, y_train)
print("Правильность на тестовом наборе: {:.2f}".format(ridge.score(X_test, y_test)))

```

```python
from sklearn.model_selection import cross_validate, GridSearchCV
clf = LogisticRegression (penalty = "l2")
grid_list = {"C": np.logspace(-4, 4, 10)}
grids = GridSearchCV (clf, param_grid = grid_list, scoring = "accuracy",return_train_score=True,cv=3)

# roc_auc is just for binary
grids.fit(X_train, y_train)
    
result = grids.cv_results_
print(grids.best_score_, grids.best_params_)
```

**3. Добавить полиномиальные признаки до второго порядка. Построить бинарный классификатор для каждого семейства на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
# Your code here
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
#l2-regularization

pipe = make_pipeline(StandardScaler(),PolynomialFeatures(),Ridge())
param_grid = {'polynomialfeatures__degree': [1, 2, 3],
'ridge__alpha': [0.001, 0.01, 0.1, 1, 10, 100],
             }
grid = GridSearchCV(pipe, param_grid=param_grid, cv=3, n_jobs=-1)
grid.fit(X_train, y_train)


```

```python
print("Наилучшие параметры: {}".format(grid.best_params_))
print("Правильность на тестовом наборе: {:.2f}".format(grid.score(X_test, y_test)))
```
