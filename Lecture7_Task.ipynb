**1. Провести интервальную оценку среднего значения признака "MFCCs_ 2" с помощью бутстрапа с точностью 95%.**

```python
# Your code here

def bootstrap(data):
    idx = np.random.randint(0, len(data), len(data))
    print(idx," idx ")
    return data[idx]


mfcc_2 = []
data = df["MFCCs_ 2"].values

for i in range(50):
    t = bootstrap(data)
    print(t," t ")
    mfcc_2.append(t.mean())

```

```python
print(np.percentile(mfcc_2, [5.0, 95.0]))
```

**2. Построить лес случайных деревьев, для определения семейства Dendrobatidae. Вывести 5 самых важных признаков. Сравнить результат с логистической регрессией.**

```python
df.iloc[:,0:22].head()
```

```python
t=(df['Family'] == 'Dendrobatidae').values
```

```python
%time
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X=df.iloc[:,0:22]
y=t

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
random_state=42)
forest = RandomForestClassifier(50,n_jobs=3)
forest.fit(X_train, y_train)
```

```python

print(forest.feature_importances_)
forest.feature_importances_
imp=forest.feature_importances_
```

```python
s=np.sort(imp)
print(s)
```

```python
i=len(s)-1
while (i>(len(s)-6)):
    l=s[i]
    idx=np.where(imp==l)
    print(s[i],idx,df.columns[idx])
    
    i-=1
```

делала отбор признаков

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

prob = forest.predict_proba(X_test)
pred = forest.predict(X_test)

print("Acc = ", accuracy_score(y_test, pred))
print("F1 = ", f1_score(y_test, pred))
print("ROC AUC = ", roc_auc_score(y_test, prob[:, 1]))
```

впечатляет

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

prob = logreg.predict_proba(X_test)
pred = logreg.predict(X_test)

print("Acc = ", accuracy_score(y_test, pred))
print("F1 = ", f1_score(y_test, pred))
print("ROC AUC = ", roc_auc_score(y_test, prob[:, 1]))
```

**3. Использовать бэггинг для обучения 100 логистических регрессий для определения семейства Dendrobatidae. Использовать RandomizedSearchCV для того, чтобы не перебирать все возможные комбинации параметров. Выбрать для RandomizedSearchCV - 20 итераций.**

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV

parameters = {
    'max_features': [5, 10, 15],    # кол-во случайно отбираемых признаков
    'max_samples': [0.55, 0.75, 0.9], # процент выборки для обучения 
    'base_estimator__C': np.logspace(-4, 5, 10), # регуляризация
}


# Your code here

clf = BaggingClassifier(LogisticRegression(solver="liblinear"), 100)

rcv = RandomizedSearchCV(clf, parameters, n_iter=20, scoring="roc_auc", cv=2)

rcv.fit(X_train, y_train)
```

```python
pred = rcv.predict(X_test)
print("Acc = ", accuracy_score(y_test, pred))
```

```python
odin=LogisticRegression(solver="liblinear")
odin.fit(X_train, y_train)
pred = odin.predict(X_test)
print("Acc = ", accuracy_score(y_test, pred))
```
